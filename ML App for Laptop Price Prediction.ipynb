```python
df1["Memory"] = df1["Memory"].astype(str).replace('\.0','',regex=True)
df1["Memory"] = df1["Memory"].str.replace('GB','')
df1["Memory"] = df1["Memory"].str.replace('TB','000')

new = df1["Memory"].str.split('+',n=1,expand=True)
df1['first']=new[0]

df1['lyer1HDD'] = df1['first'].apply(lambda x: 1 if "HDD" in x else 0)
df1['lyer1SSD'] = df1['first'].apply(lambda x: 1 if "SSD" in x else 0)
df1['lyer1Hybrid'] = df1['first'].apply(lambda x: 1 if "Hybrid" in x else 0)
df1['lyer1Flash'] = df1['first'].apply(lambda x: 1 if "Flash Storage" in x else 0)

df1['second']=new[1]
df1['second'].fillna('0',inplace=True)
df1['lyer2HDD'] = df1['second'].apply(lambda x: 1 if "HDD" in x else 0)
df1['lyer2SSD'] = df1['second'].apply(lambda x: 1 if "SSD" in x else 0)
df1['lyer2Hybrid'] = df1['second'].apply(lambda x: 1 if "Hybrid" in x else 0)
df1['lyer2Flash'] = df1['second'].apply(lambda x: 1 if "Flash Storage" in x else 0)

df1['first'] = df1['first'].str.replace(r'\D','')
df1['second'] = df1['second'].str.replace(r'\D','')
df1['first'] = df1['first'].astype(int)
df1['second'] = df1['second'].astype(int)

df1["HDD"]=(df1["first"]*df1["lyer1HDD"]+df1["second"]*df1["lyer2HDD"])
df1["SSD"]=(df1["first"]*df1["lyer1SSD"]+df1["second"]*df1["lyer2SSD"])
df1["Hybrid"]=(df1["first"]*df1["lyer1Hybrid"]+df1["second"]*df1["lyer2Hybrid"])
df1["Flash_Storage"]=(df1["first"]*df1["lyer1Flash"]+df1["second"]*df1["lyer2Flash"])

df1.drop(columns=['first', 'second', 'lyer1HDD', 'lyer1SSD', 'lyer1Hybrid',
       'lyer1Flash', 'lyer2HDD', 'lyer2SSD', 'lyer2Hybrid',
       'lyer2Flash'],inplace=True)
df1
```

# Machine Learning Process

# Consider the ip & op col for dataset

```python
x=df1.drop(columns=['Price'])#ip
y=df1['Price']#op

```

```python
df1.to_csv('laptop_price.csv')
```

```python
#divide DataSet into train & tst dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2)
print('x train',len(x_train))
print('x test',len(x_test))
print('y train',len(y_train))
print('y test',len(y_test))
```


# label encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error
step1=ColumnTransformer(transformers=[('clm',OneHotEncoder(sparse=False),[0,1,6,7,8])],remainder='passthrough')
step2=LinearRegression()
pipe=Pipeline([('step1',step1),('step2',step2)])

#model building
model=pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print("r2 score = ",r2_score(y_pred,y_test))
print("MAE = ",mean_absolute_error(y_pred,y_test))
```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

lm = LinearRegression()
knn = KNeighborsRegressor(n_neighbors=3)
dtr = DecisionTreeRegressor()
rfr =RandomForestRegressor(n_estimators=100,random_state=3,max_depth=15)
svr = SVR(kernel='rbf',C=100,epsilon=0.1)#to inc accu constant params

step2=VotingRegressor([('lm',lm),('knn',knn),('dtr',dtr),('rfr',rfr),('svr',svr)],weights=[5,1,1,1,1])

pipe=Pipeline([('step1',step1),('step2',step2)])

#model building
model=pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
y_pred
print("r2 score = ",r2_score(y_pred,y_test))
print("MAE = ",mean_absolute_error(y_pred,y_test))
```
